```python
import pandas as pd
```

```python
stories = pd.read_csv("stories.csv")
```

```python
stories = stories.drop(columns=['license_type','hi_res_downloads', 'epub_downloads', 'donor_updated', 'attribution_text', 'credit_line', 'publisher',  'description', 'illustrator_name_updated', 'author_name_updated', 'translator_name_updated', 'status', 'reads', 'downloads_updated', 'published_date_updated', 'ancestry_updated', 'derivation_type_updated'])
```

```python
stories = stories[(stories['story_langugage']=="English") | (stories['story_langugage']=="Hindi")]
```

```python
stories.columns
```

```python
tags_list = ['<p>' ,'</p>' , '<p*>','\\n', '\\t', ' +'
             '<ul>','</ul>',
             '<li>','</li>',
             '<br>',
             '<strong>','</strong>',
             '<span*>','</span>',
             '<a href*>','</a>',
             '<em>','</em>']

for tag in tags_list:
    stories['content'].replace(to_replace=tag, value='', regex=True, inplace=True)
```

```python
merged_stories = pd.merge(stories, stories[['content', 'title', 'synopsis', 'story_langugage']], right_on='title', left_on='story_original_title', suffixes=['','_original']).sort_index()
```

```python
merged_stories
```

```python
merged_stories = merged_stories.set_index('story_id')
```

```python
merged_stories = merged_stories[merged_stories.pipe(lambda x: x['story_langugage'] != x['story_langugage_original']) & merged_stories.content.notnull()]
```

```python
merged_stories[['content', 'content_original', 'story_langugage', 'story_langugage_original' ]]
```

```python
en_hn = merged_stories[merged_stories['story_langugage'] == 'English']
en_hn['content'].to_csv('pb_en-hn.en', sep=' ', index=False, header=False, mode = 'a')
en_hn['content_original'].to_csv('pb_en-hn.hn', sep=' ', index=False, header=False, mode = 'a')
```

```python
en_hn = merged_stories[merged_stories['story_langugage'] == 'Hindi']
en_hn['content'].to_csv('pb_en-hn.hn', sep=' ', index=False, header=False, mode = 'a')
en_hn['content_original'].to_csv('pb_en-hn.en', sep=' ', index=False, header=False, mode = 'a')
```
